```python
# Don't edit this cell, this create your playlist
import random,string
songids=['SONGID'+str(i).zfill(2) for i in range(1,21)]
userid=[''.join('USER')+str(random.randint(10,99)).zfill(2) for i in range(100)]
playlist={i:j for i,j in zip(userid,[random.sample(songids,random.randint(4,7)) for i in range(len(userid))])}
playlist
```

```python
myplist=['SONGID01','SONGID06','SONGID16']
```

## Find a playlist in the dictionary, where ALL of these songs are present (there can be more than 1 such playlist)

```python
## We first need to run each song from myplist so we need to cross verify by each index
# we can assign a variable s to take value of the length of myplist from 0 using range & len functions
for s in range(0,len(myplist)):
# storing each index in a variable called songs
    songs = myplist[s]
# since the playlist is a dictionary, we have to match each song against the value for each key and find the user/ playlist
    matched_playlist = ([key for key, value in playlist.items() if songs in value])
matched_playlist
```

```python
## Finding the list of songs from the matched playlists
matched_songs_list = []
for i in range(0,len(matched_playlist)):
    song_list = matched_playlist[i]
    matched_songs_list.extend([value for key, value in playlist.items() if song_list in key])
matched_songs_list
```

## From this playlist(s), pick the remaining songs and recommend to the new user (refer below examples for more details)

```python
from functools import reduce
```

```python
## arranging the matched songs

arranged_plist = reduce(lambda xs, ys: xs + ys, matched_songs_list)

## Shortlisting unique songs from this list

unique_songs = []
for i in arranged_plist:
    if i not in unique_songs:
        unique_songs.append(i)
        
## removing the existing songs in myplist from this unique songs list to recommend

recommended_songs = [arranged_plist for arranged_plist in unique_songs if arranged_plist not in myplist]
recommended_songs
```
